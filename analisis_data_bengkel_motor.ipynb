```python
import pandas as pd

# Memuat dataset
df = pd.read_csv('ledengmotor_dataset.csv')

# Menampilkan 5 baris pertama dari dataset
df.head()
```

```python
# Menampilkan ringkasan statistik
df.describe(include='all')
```

```python
# Menampilkan informasi umum tentang dataset
df.info()
```

```python
# Menampilkan distribusi nilai untuk kolom 'Jenis_Service'
df['Jenis_Service'].value_counts()
```

# Memeriksa nilai yang hilang
df.isnull().sum()


```python
# Mengisi atau menghapus nilai yang hilang
df['Model_Motor'] = df['Model_Motor'].fillna('Unknown')
```

```python
# Mengonversi kolom 'Tanggal_Service' ke tipe datetime
df['Tanggal_Service'] = pd.to_datetime(df['Tanggal_Service'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi distribusi biaya
plt.figure(figsize=(10, 6))
sns.histplot(df['Biaya'], bins=30, kde=True)
plt.title('Distribusi Biaya Service')
plt.xlabel('Biaya')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Analisis hubungan antara 'Jenis_Service' dan 'Biaya'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Jenis_Service', y='Biaya', data=df)
plt.title('Boxplot Biaya Berdasarkan Jenis Service')
plt.xlabel('Jenis Service')
plt.ylabel('Biaya')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analisis jumlah layanan berdasarkan tahun kendaraan
plt.figure(figsize=(10, 6))
sns.countplot(x='Tahun_Kendaraan', data=df)
plt.title('Jumlah Kendaraan Berdasarkan Tahun')
plt.xlabel('Tahun Kendaraan')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```
